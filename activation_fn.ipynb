Step Function:

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
x=np.arange(-10,10,0.5)

```

```python
def step_fn(n):
    if n>0:
        return 1
    if n<=0:
        return 0
```

```python
y=[]
for n in x:
    y.append(step_fn(n))
```

```python
plt.plot(x,y)
plt.grid()
```

```python
import math
def sigmoid_fn(x):
    return 1/(1+math.exp(-x))
```

```python
y=[]
for i in x:
    y.append(sigmoid_fn(i))
```

```python
plt.plot(x,y)
```

```python
def swish_fn(x):
    return 1/(1+math.exp(x))
```

```python
y=[]
for i in x:
    y.append(swish_fn(i))
```

```python
plt.plot(x,y)
plt.grid()
```
