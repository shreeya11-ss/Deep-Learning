```python
import cv2
import os
import time
```

```python
image_folder='photos\\train\\left_hand\\image'

#initialise webcam
cap=cv2.VideoCapture(0)

photo_count=1
for i in range(100):
    ret,frame=cap.read() #read the frame from webcam


    #resize the frame
    resized_frame=cv2.resize(frame,(224,224))
    image_path=image_folder+str(photo_count)+".jpg"
    cv2.imwrite(image_path,resized_frame)
    photo_count+=1
    print(f"photo saved:{photo_count}")
    time.sleep(0.2)
```

```python

cap.release()
```
